```python
#WAP that arrange all odd numbers on left side of the list and even numbers on the right side of the list.

L1=eval(input("Enter a list:"))
n=int(input("Enter the no. of times you want to shift the elements:"))
length=len(L1)
for i in range(n):
    last=L1[length-1]
    for j in range(length-1,0,-1):
        L1[j]=L1[j-1]
    L1[0]=last
print(L1)
```
